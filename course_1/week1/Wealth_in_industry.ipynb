## Wealth generated in USA industries 

I found it interesting to see how our wealth would have evolved in time had we invested in different industries.

```python
import pandas as pd
import numpy as np
import edhec_risk_kit as erk
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
ind = erk.get_ind_returns()

ind.head()
```

Let's use the function we created to plot the wealth with time.

```python
wealth_index,final_wealth = erk.wealth_index(ind, final_wealth=True)

wealth_index.plot(figsize=(18,10), legend = 2)

# wealth_index[["Food","Smoke","Oil"]].plot(figsize=(18,10), legend = 2) if we want to have a look at specific industries
```

And now let's have a look at the most lucrative industries at present time, had we started investing 1000 dollars in 1929:

```python
final_wealth.sort_values(by=["Final wealth"],ascending=False)
```

Surprisingly (or not so much), smoking industry, beer and healthcare have been the industries with bigger wealth creation.
